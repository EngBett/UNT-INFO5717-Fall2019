## Part one: Data Visualization Introduction

### What is Data Visualization?
Data (or information) visualization is used to interpret and gain insight into large amounts of data. This is achieved through visual representations, often interactive, of raw data.

### Why we need Data Visualization?

For better communication!

### How to conduct data visualization?

- Basic Consideration:

(1) Determine what you are ultimately trying to communicate
(2) Determine which visualization best achieves your aim in the clearest manner possible

- Further Considerations to create your data visualization

(1) What data (or subset of data) is relevant for your aim?

(2) Who is your audience?

(3) How will you encode your data?

(4) How will you structure the visualization?

(5) How will you demonstrate the relationship between data?

(6) Determine the scale of your visualization.

(7) Which elements will the user interact with?

(8) Will a user intuitively understand how the visualization works and what it represents?

### How to choose a chart?

- Simple comparisons: bars and columns
Applying the perceptual hierarchy of visual cues, bar and column charts are usually the best options for simple comparisons. Vertical columns often work well when few items are being compared, while horizontal bars may be a better option when there are many items to compare. As showing in the folowing example:

```python
Image('http://online.wsj.com/media/top50.png')
```

- Comparisons: change over time
Bar or column charts can also be used to illustrate change over time, but there are other possibilities, as shown in these charts showing participation in the federal government’s food stamps nutritional assistance program, from 1969 to 2014.

Each of these charts communicates the same basic information with a subtly different emphasis. The column chart emphasizes each year as a discrete point in time, while the line chart focuses on the overall trend or trajectory. The dot-and-line chart is a compromise between these two approaches, showing the trend while also drawing attention to the value for each year. (The dot-column chart is an unusual variant of a column chart, included here to show another possible design approach.)

### Other Principles
- Using color effectively
Color falls low on the perceptual hierarchy of visual cues, but as we have seen above, it is often deployed to highlight particular elements of a chart, and sometimes to encode data values. Poor choice of color schemes is a problem that bedevils many news graphics, so it is worth taking some time to consider how to use color to maximum effect.

It helps to think about colors in terms of the color wheel, which places colors that “harmonize” well together side by side, and arranges those that have strong visual contrast — blue and orange, for instance — at opposite sides of the circle:

```python
Image("https://paldhous.github.io/ucb/2016/dataviz/img/class2_18.jpg")
```

When encoding data with color, take care to fit the color scheme to your data, and the story you’re aiming to tell. Color is often used to encode the values of categorical data. Here you want to use “qualitative” color schemes, where the aim is to pick colors that will be maximally distinctive, as widely spread around the color wheel as possible:

```python
Image("https://paldhous.github.io/ucb/2016/dataviz/img/class2_19.jpg")
```

When using color to encode continuous data, it usually makes sense to use increasing intensity, or saturation of color to indicate larger values. These are called “sequential” color schemes:

```python
Image("https://paldhous.github.io/ucb/2016/dataviz/img/class2_20.jpg")
```

In some circumstances, you may have data that has positive and negative values, or which highlights deviation from a central value. Here, you should use a “diverging” color scheme, which will usually have two colors reasonably well separated on the color wheel as its end points, and cycle through a neutral color in the middle:

```python
Image("https://paldhous.github.io/ucb/2016/dataviz/img/class2_21.jpg")
```

- Using chart furniture, minimizing chart junk, highlighting the story

(1) Title and subtitle These provide context for the chart.

(2) Coordinate system For most charts, this is provided by the horizontal and vertical axes, giving a cartesian system defined by X and Y coordinates; for a pie chart it is provided by angles around a circle, called a polar coordinate system.

(3) Scale Labeled tick marks and grid lines can help your audience read data values.

(4) Labels You will usually want to label each axis. Think about other labels that may be necessary to explain the message of your graphic.

(5) Legend If you use color or shape to encode data, you will often need a legend to explain this encoding.

(6) Source information Usually given as a footnote. Don’t forget this!

- Histogram
A histogram takes in a series of data and divides the data into a number of bins. It then plots the frequency data points in each bin (i.e. the interval of points). It is useful in understanding the count of data ranges.

When to use: We should use histogram when we need the count of the variable in a plot.

Components of a histogram:

(1) n: Contains the frequency of each bin

(2) bins: Represents the middle value of each bin

(3) patches: The Patch object for the rectangle shape representing each bar

```python
# Example 1: Number of particular games sold in a store.

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

# Preparing the data
np_data=pd.read_csv(r'D:\Academic_Study\UNT_PHD\Courses\INFO5717\2019fall\Lesson_10_data\national_parks.csv')
print(np_data.head())

# Histogram

plt.hist(np_data['GrandCanyon'], 
         facecolor='peru', 
         edgecolor='blue', 
         bins=10)
plt.show()
```

```python
# Example 2: Number of particular games sold in a store.

# Multiple histograms
plt.hist(np_data['BryceCanyon'], 
         facecolor='orangered', 
         edgecolor='maroon', 
         bins=8)

plt.hist(np_data['GrandCanyon'], 
         facecolor='peru', 
         edgecolor='blue', 
         bins=10, 
         alpha = 0.3)

plt.show()
```

- Time Series by line plot
Time series is a line plot and it is basically connecting data points with a straight line. It is useful in understanding the trend over time. It can explain the correlation between points by the trend. An upward trend means positive correlation and downward trend means a negative correlation. It mostly used in forecasting, monitoring models.

When to use: Time Series should be used when single or multiple variables are to be plotted over time.

```python
# Example 4: Stock Market Analysis of Companies

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Preparing the data
stock_data = pd.read_csv(r'D:\Academic_Study\UNT_PHD\Courses\INFO5717\2019fall\Lesson_10_data\stocks.csv')
print(stock_data.head())
stock_data['Date'] = pd.to_datetime(stock_data['Date'])
print(stock_data.head())

# Time Series by line plot
# Figsize for width and height of plot
fig = plt.figure(figsize=(15,7))

fig.suptitle('Stock price comparison 2007-2017', 
             fontsize=20)

ax1 = fig.add_subplot(231)
ax1.set_title('MSFT')

ax1.plot(stock_data['Date'],
         stock_data['MSFT'], 
         color='green')

ax2 = fig.add_subplot(232)
ax2.set_title('GOOG')

ax2.plot(stock_data['Date'],
         stock_data['GOOG'], 
         color='purple')

ax3 = fig.add_subplot(233)
ax3.set_title('SBUX')

ax3.plot(stock_data['Date'],
         stock_data['SBUX'], 
         color='magenta')

ax4 = fig.add_subplot(234)
ax4.set_title('ADBE')

ax4.plot(stock_data['Date'],
         stock_data['ADBE'], 
         color='orange')

ax4 = fig.add_subplot(235)
ax4.set_title('NFLX')

ax4.plot(stock_data['Date'],
         stock_data['NFLX'], 
         color='chocolate')

ax4 = fig.add_subplot(236)
ax4.set_title('ORCL')

ax4.plot(stock_data['Date'],
         stock_data['ORCL'], 
         color='teal')

plt.show()
```

```python
# Example 5: Credit Score of Customer

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

# Preparing the data
exam_data = pd.read_csv(r'D:\Academic_Study\UNT_PHD\Courses\INFO5717\2019fall\Lesson_10_data\exams.csv')
exam_scores = exam_data[['math score', 'reading score', 'writing score']]
print(exam_scores.head())
print(exam_scores.describe())

# For plotting in boxplot, we convert it to an array
exam_scores_array = np.array(exam_scores)

# Box plots
colors = ['blue', 'grey', 'lawngreen']

bp = plt.boxplot(exam_scores_array, 
                 patch_artist=True,
                 notch=True)

for i in range(len(bp['boxes'])):
    
    bp['boxes'][i].set(facecolor=colors[i])
    
    bp['caps'][2*i + 1].set(color=colors[i])
    
plt.xticks([1, 2, 3], ['Math', 'Reading', 'Writing'])

plt.show()
```

- Stack Plot and Stem Plot

Stack plot visualizes data in stacks and shows the distribution of data over time.

When to use: It is used for checking multiple variable area plots in a single plot.

```python
# Example 6: national_parks 

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Preparing the data
np_data= pd.read_csv(r'D:\Academic_Study\UNT_PHD\Courses\INFO5717\2019fall\Lesson_10_data\national_parks.csv')
np_data.head()

x = np_data['Year']
y = np.vstack([np_data['Badlands'], 
               np_data['GrandCanyon'], 
               np_data['BryceCanyon']])

# Stack Plots: visualize data overtime
# Labels for each stack
labels = ['Badlands', 
          'GrandCanyon', 
          'BryceCanyon']

# Colors for each stack
colors = ['sandybrown', 
          'tomato', 
          'skyblue']

#  Similar to pandas df.plot.area()
plt.stackplot(x, y, 
              labels=labels,
              colors=colors, 
              edgecolor='black')

# Plots legend to the upperleft of Figure
plt.legend(loc=2)

plt.show()
```

```python
# Example 7: national_parks 

np_stem= np_data.copy()
np_stem[['Badlands', 
         'GrandCanyon',
         'BryceCanyon']] = np_data[['Badlands', 
                                    'GrandCanyon',
                                    'BryceCanyon']].diff()

print(np_stem.head())

# Stem Plots: Similar to Stack Plots but it helps us view negative numbers, i.e the difference in data over time.
plt.figure(figsize=(15,11))

plt.suptitle('Change in Number of Visitors', y=0.94)

plt.subplot(311)
plt.stem(np_stem['Year'],
         np_stem['Badlands'],
         markerfmt = 'b_',
         linefmt = 'r--',
         basefmt = 'g:')
plt.title('GrandCanyon')

plt.subplot(312)
plt.stem(np_stem['Year'],
         np_stem['GrandCanyon'],
         markerfmt = 'b_',
         linefmt = 'r--',
         basefmt = 'g:')
plt.title('Badlands')

plt.subplot(313)
plt.stem(np_stem['Year'],
         np_stem['BryceCanyon'],
         markerfmt = 'b_',
         linefmt = 'r--',
         basefmt = 'g:')
plt.title('BryceCanyon')

plt.show()
```

- Bar Plot
Bar Plot shows the distribution of data over several groups. It is commonly confused with a histogram which only takes numerical data for plotting. It helps in comparing multiple numeric values.

When to use: It is used when to compare between several groups.

```python
# Example 8: Student marks in an exam.

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

countries=pd.read_csv(r'D:\Academic_Study\UNT_PHD\Courses\INFO5717\2019fall\Lesson_10_data\countries.csv')
data_2007=countries[countries['year']==2007]
top_10_p=data_2007.sort_values('population',ascending=False).head(10)
print(top_10_p)

plt.figure(figsize=(12,6))

x=range(10)
plt.bar(x,top_10_p['population']/10**9)
plt.xticks(x,top_10_p['country'])
plt.xlabel('Countries')
plt.ylabel('Population in Billions')
plt.title('10 Most Populous Countries in 2007')
plt.show()
```

- Scatter Plot
Scatter plot helps in visualizing 2 numeric variables. It helps in identifying the relationship of the data with each variable i.e correlation or trend patterns. It also helps in detecting outliers in the plot.

When to use: It is used in Machine learning concepts like regression, where x and y are continuous variables. It is also used in clustering scatters or outlier detection.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

countries=pd.read_csv(r'D:\Academic_Study\UNT_PHD\Courses\INFO5717\2019fall\Lesson_10_data\countries.csv')
countries.head()
data_2007=countries[countries['year']==2007]
print(data_2007)
plt.scatter(data_2007['gdpPerCapita'],data_2007['lifeExpectancy'])
plt.title('Life Expectancy vs GDP Per Capita in 2007 ')
plt.xlabel('GDP per Capita($)')
plt.ylabel('Life Expectancy')
plt.show()
```

- 3D Scatterplot

3D Scatterplot helps in visualizing 3 numerical variables in a three- dimensional plot.

```python
np.log10(data_2007['gdpPerCapita']).corr(data_2007['lifeExpectancy'])
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(countries['year'],countries['gdpPerCapita'],countries['lifeExpectancy'])
plt.show()
```
